```python
import torch
import random
import time
```

```python
def random_generate_data(positvie_num_need, negative_num_need):
  # line equation y=mx+b
  m = 0.7
  b = 0.3

  # Generate 30 random 2D data samples
  data = []
  current_positive_num = 0
  current_negative_num = 0

  while len(data) < positvie_num_need + negative_num_need:
      x = random.uniform(-10, 10) # Uniformly generate a random x value between -10 and 10
      y = random.uniform(-10, 10) # Uniformly generate a random y value between -10 and 10
      w = torch.tensor([m, -1, b]) # mx - y + b = 0
      X = torch.tensor([x, y, 1])
      label = -1
      # mark the sample as positive (1) or negative (-1)
      if torch.dot(w, X) > 0 and current_positive_num < positvie_num_need:
        label = 1
        current_positive_num += 1
        # data.append(torch.tensor([x, y, label]))
        data.append([x, y, label])
      elif torch.dot(w, X) < 0 and current_negative_num < negative_num_need:
        label = -1
        current_negative_num += 1
        # data.append(torch.tensor([x, y, label]))
        data.append([x, y, label])
  
  # Shuffle the data
  random.shuffle(data)

  # data = torch.tensor(data)

  # Split the data into positive and negative samples
  # [sample[:2] for sample in data if sample[2] == 1][:15] is a list comprehension in Python 
  # that extracts the first two elements of the tuples in the data list where the third element 
  # (i.e., sample[2]) is equal to 1 (i.e., positive samples), and then selects the first 
  # 15 elements of the resulting list.
  positive_samples = [sample[:3] for sample in data if sample[2] == 1][:positvie_num_need]
  negative_samples = [sample[:3] for sample in data if sample[2] == -1][:negative_num_need]

  # Print the positive and negative samples
  print("Length of data:", len(data))
  print("Number of positive samples: ", len(positive_samples))
  print("Number of negative samples: ", len(negative_samples))
  print("Positive samples:" + str(positive_samples))
  print("Negative samples:" + str(negative_samples))
  print("Data:", data)

  return data
```

```python
random_generate_data(15, 15)
```

```python
def PLA(data,w0):
  w = w0 
  iterations = 0
  while True:
    num_errors = 0
    for elements in data:
      x, y, label = elements
      if compare(w, x, y, label): #check if less than zero
        w = w + label * torch.tensor([x, y, 1])
        num_errors += 1
    iterations += 1
    if num_errors == 0:
      break
  return w, iterations
```

```python
def compare(w, x, y, label):
  return True if label * torch.dot(w,torch.tensor([x, y, 1], dtype=torch.float32)) <= 0 else False
```

```python
w0 = torch.tensor([1, 1, 1], dtype=torch.float32)
data = random_generate_data(15, 15)
data = torch.tensor(data) #Set to tensor
w, iterations = PLA(data, w0)
print("Number of iterations:", iterations)
print("Final weight:", w)
```

```python
def error_rate(data, w):

  num_errors = 0

  for elements in data:
    x, y, label = elements
    if compare(w, x, y, label): #check if less than zero
      w = w + label * torch.tensor([x, y, 1])
      num_errors += 1

    
  return num_errors / len(data), w
```

```python
# Pocket Algorithm
def pocket_algorithm(data, num_iterations):
  w_current = torch.tensor([0, 0, 0], dtype=torch.float32)
  w_pocket = torch.tensor([0, 0, 0], dtype=torch.float32)
  error_current = 0
  error_pocket = 1
  # random_data_points = random.sample(data, random.randint(1, len(data))) 
  random_data_points = data[:random.randint(1, len(data))] # Random select unspecific amount of data
  # random_data_points = data
  for i in range(num_iterations):
    error_current, w_current= error_rate(random_data_points, w_current)
    if error_current <= error_pocket:
      error_pocket = error_current
      w_pocket = w_current

  return w_pocket, error_pocket

```
